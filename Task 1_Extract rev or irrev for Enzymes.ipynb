```python
# Make matching pattern
from nltk.stem.snowball import SnowballStemmer
import spacy
from spacy.matcher import Matcher 
from spacy import displacy 
import visualise_spacy_tree
from IPython.display import Image, display

# load english language model
nlp = spacy.load('en_core_web_sm',disable=['ner','textcat'])

stemmer = SnowballStemmer("english")

def find_rev_or_irrev(enzyme_name, text):
    
    enz_reaction = []
    sent = []
    
    doc = nlp(text)
    
    # pattern WITH ORDER OF PRIORITY - WITH A NOUN HAS HIGHEST PRIORITY (BREAK INCLUDE)
    pattern = [{'POS':'PRON', 'OP':'+'},
               {'POS':'NOUN', 'OP':'+'},
               {'POS':'ADV', 'OP':'?'},
               {'POS':'VERB', 'OP':'?'},
               {'POS':'ADJ', 'OP':'?'},
               {'POS':'NOUN', 'OP':'?'},
               {'POS':'DET', 'OP':'?'}]
    
    # Allow specific nouns
    noun_allowed = ['enzyme', 'Enzyme','enzymes', 'Enzymes', enzyme_name.upper(), enzyme_name.lower(), enzyme_name.capitalize()]
    
    pattern1 = [{'POS':'NOUN', 'TEXT':noun_allowed[0]},
               {'POS':'DET', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'NOUN', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'VERB', 'OP':'?'},
               {'POS':'ADV'},
               {'POS':'VERB'}]
    
    pattern12 = [{'POS':'NOUN', 'TEXT':noun_allowed[1]},
               {'POS':'DET', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'NOUN', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'VERB', 'OP':'?'},
               {'POS':'ADV'},
               {'POS':'VERB'}]
    
    pattern13 = [{'POS':'NOUN', 'TEXT':noun_allowed[2]},
               {'POS':'DET', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'NOUN', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'VERB', 'OP':'?'},
               {'POS':'ADV'},
               {'POS':'VERB'}]
    
    pattern14 = [{'POS':'NOUN', 'TEXT':noun_allowed[3]},
               {'POS':'DET', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'NOUN', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'VERB', 'OP':'?'},
               {'POS':'ADV'},
               {'POS':'VERB'}]
    
    pattern15 = [{'POS':'NOUN', 'TEXT':noun_allowed[4]},
               {'POS':'DET', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'NOUN', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'VERB', 'OP':'?'},
               {'POS':'ADV'},
               {'POS':'VERB'}]
    
    pattern16 = [{'POS':'NOUN', 'TEXT':noun_allowed[5]},
               {'POS':'DET', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'NOUN', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'VERB', 'OP':'?'},
               {'POS':'ADV'},
               {'POS':'VERB'}]
    
    pattern17 = [{'POS':'NOUN', 'TEXT':noun_allowed[6]},
               {'POS':'DET', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'NOUN', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'VERB', 'OP':'?'},
               {'POS':'ADV'},
               {'POS':'VERB'}]
    
    
    pattern2 = [{'POS':'NOUN'},
               {'POS':'VERB', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'NOUN', 'OP':'?'},
               {'POS':'VERB', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},               
               {'POS':'ADJ'},
               {'POS':'NOUN'}]
    
    
    pattern3 = [{'POS':'NOUN','OP':'?'},
               {'POS':'ADJ', 'OP':'?'},
               {'POS':'DET', 'OP':'?'},
               {'POS':'AUX', 'OP':'?'},
               {'POS':'VERB'},
               {'POS':'DET', 'OP':'?'},  
               {'POS':'DET', 'OP':'?'}, 
               {'POS':'DET', 'OP':'?'},  
               {'POS':'ADJ'}]
    
    
    # match all 3 patterns and append to list
    enzyme_reaction = {}
    enzyme_reaction[enzyme_name] = []
    reaction_values = []
    
    try:
        PATTERNS = [pattern1, pattern12, pattern13, pattern14, pattern15, pattern16, pattern17, pattern2, pattern3]
        for j in range(len(PATTERNS)):
            # Matcher class object 
            matcher = Matcher(nlp.vocab) 
            matcher.add("matching_" + str(j), [PATTERNS[j]], on_match=None) 

            matches = matcher(doc)
            flag=0

            for i in range(0,len(matches)):

                # match: id, start, end
                start, end = matches[i][1], matches[i][2]

                # matched string
                span = str(doc[start:end]).lower()

                if ( span ):
                    ps_span = [stemmer.stem(word) for word in span.split(" ")]
                    if( 'revers' in ps_span and 'Reversible' not in reaction_values ):
                        reaction_values.append('Reversible')
                        flag=1
                        break
                    if( 'irrevers' in ps_span and 'Irreversible' not in reaction_values):
                        reaction_values.append('Irreversible')
                        flag=1
                        break

            if(flag == 1):
                break

        enzyme_reaction[enzyme_name] = reaction_values 
        
    except:
        pass
#         for node in token.subtree:
            
            # only extract the proper nouns
#             if ( (node.pos_ == 'ADJ' and stemmer.stem(node.text) in ['revers', 'irrevers']) or  (node.pos_ == 'ADV' and stemmer.stem(node.text) in ['revers', 'irrevers']) ):
#                 op_text = ' '.join(str(node) for node in token.subtree)
#                 print(op_text, node.pos_)
#                 enz_reaction[enzyme_name] = (op_text, node.text)
#                 flag = 1
#         if(flag==1):
#             break

            
            
# fp on alcohol dehydrogenases      
    return enzyme_reaction
```

```python
# make this as csv
import pandas as pd
enzyme_df = pd.DataFrame()

for i in range(len(final_enz)):
    if(len(final_enz[i]) > 0):
        id_val = ''
        de_val = ''
        an_val = ''
        
        for j in range(len(final_enz[i])):
            if(final_enz[i][j][:2] == 'ID'):
                id_val = final_enz[i][j].split('ID')[1].strip()
            elif(final_enz[i][j][:2] == 'DE'):
                de_val = final_enz[i][j].split('DE')[1].strip()
            elif(final_enz[i][j][:2] == 'AN'):
                an_val = final_enz[i][j].split('AN')[1].strip()
            
            enz_name = ''
            enz_name_an = ''
            if(de_val):
                enz_name = de_val + ' (EC ' + id_val + ')'
            if(an_val):
                enz_name_an = an_val + ' (EC ' + id_val + ')'
            
        enzyme_df = enzyme_df.append([{'Enzyme_name':enz_name, 'Alternate Name':enz_name_an, 'Reversible_or_Irreversible': False}], ignore_index = True)
                
```
